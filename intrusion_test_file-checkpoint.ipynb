```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
pd.set_option("display.max_columns",None)
import numpy as np
import pickle
```

```python
class_labels={0:'BENIGN', 1:'DDoS'}
```

```python
with open(file="models/RandomForestClassifier.pkl",mode="rb") as file:
    model=pickle.load(file=file)
```

```python
with open(file="models/Normalization_model.pkl",mode="rb") as file:
    scaler=pickle.load(file=file)
```

```python
input_df=pd.read_excel("test_excel.xlsx")
input_df.head(100)
```

```python
scaled_df=scaler.transform(input_df)
final_df=pd.DataFrame(data=scaled_df,columns=input_df.columns)
final_df.head()
```

```python
model_prediction=model.predict(final_df)
print(model_prediction)
```

```python
string_labels=[]
for num_label in model_prediction:
    string_labels.append(class_labels[num_label])
print(string_labels)
```

```python
input_df['ModelPrediction']=string_labels
input_df.head(100)
```
